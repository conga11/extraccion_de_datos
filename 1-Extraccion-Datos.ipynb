```python
import pandas as pd
```

```python
# Lee el excel
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
```

```python
df.info()
```

```python
df.head(2)
```

```python
# Filtro por objeto

filtro1 = df[df['NOMBRE_VENDEDOR'] == 'LETICIA RAMIREZ HERNANDEZ']

# Se usó la columna NOMBRE_VENDEDOR para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_objeto
filtro1.to_csv('filtro_objeto.csv')
filtro1.head(2)
```

```python
# Filtro por filas

filtro2 = df.iloc[1:4,:] # Filas consecutivas y todas las cols
filtro2 = df.iloc[[0, 2, 5, 7, 9], : ] # Filas no consecutivas y todas las cols


# Se usaron las 1 a 4 y 0, 2, 5, 7, 9 filas para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_filas
filtro2.to_csv('filtro_filas.csv')
filtro2.head(2)
```

```python
# Filtro por columnas

filtro3 = df.iloc[:, 0:3] # Todas las filas y columnas consecutivas
filtro3 = df.iloc[:, [0, 2, 5, 7, 9]] # Todas las filas y columnas no consecutivas
filtro3.head(2)

# Se usaron las columnas 0 a 3 y 0, 2, 5, 7, 9 para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_columnas

filtro3.to_csv('filtro_columnas.csv')
filtro3.head(2)
```

```python
#Filtro de filas vs columnas

filtro4 = df.iloc[1:4, 0:3] # Filas y columnas consecutivas
# LO MISMO CON NOMBRE
filtro4 = df.loc[[1,8,4], ['NOMBRE_VENDEDOR']] # Filas y columnas consecutivas


# Se usaron las filas 1 a 4 y columnas 0 a 3 para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_filas_columnas

filtro4.to_csv('filtro_filas_columnas.csv')
filtro4.head(2)

```

```python
# Filtro por cabecera

filtro5 = df.head(3)
filtro5

# Se usaron las primeras 3 filas para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_cabecera

filtro5.to_csv('filtro_cabecera.csv')
filtro5
```

```python
# Filtro por cola

filtro6 = df.tail(3)
filtro6

# Se usaron las últimas 3 filas para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_cola

filtro6.to_csv('filtro_cola.csv')
filtro6
```

```python
# Filtro por comparacion

filtro7 = df[df['PRECIO_UNITARIO'] > 1000]


# Se usaron los datos de la columna PRECIO_UNITARIO mayores a 1000 para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_comparacion

filtro7.to_csv('filtro_comparacion.csv')
filtro7.head(2)
```

```python
# Filtro por compracion con conecatores lógicos AND

filtro8 = df[(df['PRECIO_UNITARIO'] > 1000) & (df['PRECIO_UNITARIO'] < 2000)]

# Se usaron los datos de la columna PRECIO_UNITARIO mayores a 1000 y menores a 2000 para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_comparacion_conectoresAND

filtro8.to_csv('filtro_comparacion_conectoresAND.csv')
filtro8.head(2)
```

```python
# Filtro por comparacion con conecatores lógicos OR cols distintas

filtro9 = df[(df['PRECIO_UNITARIO'] > 1000) | (df['CANT'] < 4)]


# Se usaron los datos de la columna PRECIO_UNITARIO mayores a 1000 y menores a 4 para filtrar los datos

# Se guarda el archivo en formato csv con el nombre de filtro_comparacion_conectoresOR

filtro9.to_csv('filtro_comparacion_conectoresOR.csv')
filtro9
```

```python
# Filtro por comparacion con conecatores lógicos NOT

filtro10 = df[~(df['PRECIO_UNITARIO'] > 1000)]

# Se usaron los datos de la columna PRECIO_UNITARIO menores a 1000 para filtrar los datos
# Filtra los datos que no cumplen la condición y la condición es no ser mayor a 1000
filtro10.to_csv('filtro_comparacion_conectoresNOT.csv')
filtro10.head(2)
```
